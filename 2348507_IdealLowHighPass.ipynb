# **Ideal Low Pass and High Pass Filter**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fftshift, ifftshift, fft2, ifft2
from google.colab.patches import cv2_imshow
from skimage.transform import resize
```

# **Original Image**

```python
f = cv2.imread('/content/sleepyFox.jpg',0)

plt.imshow(f, cmap='gray')
plt.axis('off')
plt.show()
```

Image in frequency domain

```python
F = np.fft.fft2(f)
plt.imshow(np.log1p(np.abs(F)),cmap='gray')
plt.axis('off')
plt.show()
```

Visualizing the frequency content of the image

```python
Fshift = np.fft.fftshift(F)
plt.imshow(np.log1p(np.abs(Fshift)),cmap='gray')
plt.axis('off')
plt.show()
```

# **Low pass filter**

```python
M,N = f.shape
H = np.zeros((M,N), dtype=np.float32)
D0 = 50
for u in range(M):
    for v in range(N):
        D = np.sqrt((u-M/2)**2 + (v-N/2)**2)
        if D <= D0:
            H[u,v] = 1
        else:
            H[u,v] = 0

plt.imshow(H, cmap='gray')
plt.axis('off')
plt.show()
```

Ideal Low Pass Filter

```python
Gshift = Fshift * H
plt.imshow(np.log1p(np.abs(Gshift)),
           cmap='gray')
plt.axis('off')
plt.show()
```

Inversing the shifted Fourier Transform

```python
G = np.fft.ifftshift(Gshift)
plt.imshow(np.log1p(np.abs(G)),
           cmap='gray')
plt.axis('off')
plt.show()

g = np.abs(np.fft.ifft2(G))
plt.imshow(g, cmap='gray')
plt.axis('off')
plt.show()
```

# **High Pass Filter**

```python
H = 1 - H

plt.imshow(H, cmap='gray')
plt.axis('off')
plt.show()
```

Ideal High Pass Filtering

```python
Gshift = Fshift * H
plt.imshow(np.log1p(np.abs(Gshift)),cmap='gray')
plt.axis('off')
plt.show()
```

Inversing the fourier transform

```python
G = np.fft.ifftshift(Gshift)
plt.imshow(np.log1p(np.abs(G)),cmap='gray')
plt.axis('off')
plt.show()

g = np.abs(np.fft.ifft2(G))
plt.imshow(g, cmap='gray')
plt.axis('off')
plt.show()
```
